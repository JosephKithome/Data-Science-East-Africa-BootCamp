```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1.How do I increase the size of my figure?


```python
fig = plt.plot(figsize =(8,4))
```

```python
X = np.array([4,5,6,7,8])
y = X**2
plt.plot(X,y)
plt.ylabel('value of y')
plt.xlabel('value of x')
plt.title('value of x plotted again y')
plt.show()
```

# 2. What is a subplot? How do I create and navigate subplots?

The Matplotlib subplot () function can be called to plot two or more plots in one figure. The subplots () function takes three arguments that describes the layout of the figure

```python
X = np.array([4,5,6,7,8])
y = X**2
plt.plot(X,y, 'X',color ='red' , alpha =0.8)
plt.ylabel('value of y')
plt.xlabel('value of x')
plt.title('value of x plotted again y')
plt.show()
plt.subplot(1,2,1)

```

# 4). How do I add gridlines to my plots?



```python
x = range(1,10)
plt.plot(x,[i for i in x])
plt.plot(x,[i**2 for i in x])
plt.grid()
plt.show()
```

# 5). How do I change my line colors/width/style?



```python
x = range(0,20)
plt.plot(x,[i**3 for i in x],color ='yellow',linewidth =6)
plt.grid()
plt.show()
```

# 6.How do I change my marker style or color?

```python
x =range(1,10)
plt.plot(x,[i for i in x], '*', color ='red')
plt.plot(x,[i**2 for i in x], '^')
plt.plot(x,[i**3 for i in x], '+')
plt.grid()
plt.show()
```

# 7.How do I change the transparency for my lines?

```python
#we define the value of alpha to change the transparency of a line
fig= plt.figure()

axes= fig.add_axes([0,0,1,1])

x= np.arange(0,11)

axes.plot(x,x**2, color='red', alpha=0.8)

plt.show()
```

# 8.How do I add a legend to my plot?

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
x = np.arange(0,11)
axes.plot(x,x**2,label ='xSquared')
axes.plot(x,x**3,label ='xCubed')
axes.legend()
plt.show()
```
